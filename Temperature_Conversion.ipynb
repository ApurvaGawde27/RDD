THIS PROGRAM WILL DEMONSTRATE CREATING AN RDD FROM A LIST OF TEMPERATURES IN CELSIUS =, PERFORMING TRANSFORMATIONS TO CONVERT THEM TO FAHRENHEIT, AND ACTIONS TO COLLECT AND PRINT THE RESULTS.

```python
from pyspark import SparkConf, SparkContext
conf = SparkConf().setAppName("PySparkTemperatureConversion").setMaster("local")
sc = SparkContext(conf=conf)
```

```python
celsius_temps_apurva = [0,20,30,40,50]
rdd_apurva = sc.parallelize(celsius_temps_apurva)
```

```python
fahrenheit_rdd_apurva = rdd_apurva.map(lambda c: (c, (c * 9/5) + 32))
```

```python
fahrenheit_temps_apurva = fahrenheit_rdd_apurva.collect()
```

```python
for c, f in fahrenheit_temps_apurva:
  print(f"{c}c = {f}f")

sc.stop()
```
