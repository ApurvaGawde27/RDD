**PROGRAM 1 : WORD COUNT**

THIS PROGRAM WILL DEMONSTRATE CREATING  AN RDD FROM A TEXT DATASET, PERFORMING TRANSFORMATIONS TO COUNT THE OCCURRENCES OF EACH WORD, AND ACTIONS TO COLLECT AND PRINT THE RESULT.

```python
from pyspark import SparkConf, SparkContext
conf = SparkConf().setAppName("PySparkWordCount").setMaster("local")
sc = SparkContext(conf=conf)
```

```python
data_Apurva = ["hello world", "hello spark","hello RDD","hello PySpark","hello Apurva"]
rdd_Apurva = sc.parallelize(data_Apurva)
print(rdd_Apurva)
```

```python
words_rdd_Apurva = rdd_Apurva.flatMap(lambda sentence: sentence.split(" "))
```

```python
word_pairs_rdd_Apurva = words_rdd_Apurva.map(lambda word: (word, 1))
```

```python
word_counts_rdd_Apurva = word_pairs_rdd_Apurva.reduceByKey(lambda a, b: a+b)
```

```python
word_counts_Apurva = word_counts_rdd_Apurva.collect()
```

```python
for word, count in word_counts_Apurva :
  print(f"{word}: {count}")

sc.stop()
```
